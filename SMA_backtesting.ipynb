```python
# Import libraries
from vbt_backtesting import get_data, heatmap, SMAstrategy, Backtesting

import warnings
warnings.filterwarnings("ignore")
```

```python
# Download data
BTC_USDT = get_data()
close_price = BTC_USDT['close']
close_price.shape
```

# SMA crossover strategy

### With no commision

```python
# Check the combinations whith commission factor
  
# Define the parameter ranges for optimization
parameter_ranges = {"window_sma_1": range(1, 61), "window_sma_2": range(1, 61)}
strategy = SMAstrategy(window_sma_1=2, window_sma_2=10)
backtest = Backtesting(strategy, close_price, parameters=parameter_ranges, commission=0.00)

# Run optimization and get the best parameters
results_no_commision_SMA = backtest.run_simulation()
```

```python
# Get best returns
results_no_commision_SMA.nlargest(20, 'return_multiplier_strategy')
```

```python
heatmap(data=results_no_commision_SMA, 
        values='return_multiplier_strategy', 
        index='window_sma_1', 
        columns='window_sma_2', 
        title='Return Multipliers on SMAs lengths combinations (commision not included)', 
        size=(1000,1200))
```

### With commision

```python
# Check the combinations whith commission factor   
    
# Define the parameter ranges for optimization
parameter_ranges = {"window_sma_1": range(1, 61), "window_sma_2": range(1, 61)}
strategy = SMAstrategy(window_sma_1=2, window_sma_2=10)
backtest = Backtesting(strategy, close_price, parameters=parameter_ranges, commission=0.001)

# Run optimization and get the best parameters
results_commision_SMA = backtest.run_simulation()
```

```python
# Get best returns
results_commision_SMA.nlargest(20, 'return_multiplier_strategy')
```
