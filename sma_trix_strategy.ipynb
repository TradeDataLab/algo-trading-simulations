```python
import pandas as pd
import itertools
import tqdm 
from multiprocessing import Pool 
from defs import *
```

```python
# Create lists with ranges of SMAs values for combinations calculation
# Short SMA list
windows_sma_short = list(range(2,61))
# Long SMA list
windows_sma_long = list(range(2,61))
# Calculate all of the combinations
sma_combinations = list(itertools.product(windows_sma_short, windows_sma_long))
# Exclude pairs where length do not meat constraints
sma_combinations = [windows for windows in sma_combinations if windows[0] != windows[1]]
sma_combinations = [windows for windows in sma_combinations if windows[0] < windows[1]]
```

```python
# Simmulate SMA strategy scenarios
if __name__ == '__main__':
   with Pool() as p:
      sma_results_list = list(tqdm.tqdm(p.imap(simmulate_combinations_sma, sma_combinations), total=len(sma_combinations)))
      sma_results_df = pd.DataFrame(sma_results_list)
      
sma_results_df.nlargest(10, 'overall_profit_multiplier')   
```

```python
# List with possible TRIX window lengths
windows_trix = list(range(2,61))
```

```python
# Simmulate SMA/TRIX strategy scenarios
if __name__ == '__main__':
   with Pool() as p:
      trix_results_list = list(tqdm.tqdm(p.imap(simmulate_combinations_trix, windows_trix), total=len(windows_trix)))
      trix_results_df = pd.DataFrame(trix_results_list)
      
trix_results_df.nlargest(59, 'average_profit_multiplier')[trix_results_df['average_profit_multiplier'] > 1.001]  # Average profit multiplier higher than commission
```
